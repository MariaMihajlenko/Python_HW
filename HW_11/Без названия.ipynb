```python
from sympy import*
```

```python
x = Symbol('x')
```

```python
f,g = Symbol('f(x)'), Symbol('g(x)')
```

```python
funcF = x**3 - 50*x
```

```python
funcG = -x**4 + 88*x**2- 241
```

```python
funcE = funcF - funcG
```

```python
plot(funcF,funcG)
```

```python
plot(funcE)
```

```python
solve(funcE)
```

```python
q = solve(funcE)
```

```python
q[0] = q[0].evalf(5,chop = True)
q[1] = q[1].evalf(5,chop = True)
q[2] = q[2].evalf(5,chop = True)
q[3] = q[3].evalf(5,chop = True)

```
